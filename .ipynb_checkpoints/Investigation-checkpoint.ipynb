# Get Zip

```python
all_addresses = [(int(num),str(nm)) for _,num,nm in permits[['STREET_NUMBER','full_street_nm']].itertuples()]
```

```python
addr_hundreds_dict = {addr:set() for _,addr in all_addresses}
```

```python
for num,addr in all_addresses:
    hundred = int(num/100)*100
    addr_hundreds_dict[addr].add(hundred)
```

```python
all_addresses[0]
```

```python
addr_hundreds_df = pd.DataFrame([(num,addr) for addr,li in addr_hundreds_dict.items() for num in list(li)])
```

```python
addr_hundreds_df = addr_hundreds_df.rename({0:'street_num',1:'street_name'}, axis=1).sort_values(['street_name','street_num'])
```

```python
addr_hundreds_df.head()
```

```python
permit_attr_hundreds_dict = {str(num) + ' ' + nm:None for _,num,nm in addr_hundreds_df.itertuples()}
```

```python
restaurants = pd.read_csv('Food_Inspections.csv')
```

```python
rest_addr_zip = restaurants[['Address','Zip']][~restaurants.Address.apply(lambda s: len(str(s).split())==0)].copy()
```

```python
rest_addr_zip['street_num'] = rest_addr_zip.Address.apply(lambda s: str(s).split()[0])
```

```python
import itertools
```

```python
rest_addr_zip['street_num'] = rest_addr_zip.street_num.apply(lambda s: int("".join(itertools.takewhile(str.isdigit, s))))
```

```python
rest_addr_zip['street_name'] =  rest_addr_zip.Address.apply(lambda s: ' '.join(str(s).split()[1:]))
```

```python
rest_addr_zip.head()
```

```python
for _,_,zip_code,num,nm in rest_addr_zip.itertuples():
    hundreds = 100*int(num/100)
    simple_nm = str(hundreds) + ' ' + nm
    if simple_nm in permit_attr_hundreds_dict:
        if not pd.isna(zip_code):
            permit_attr_hundreds_dict[simple_nm] = int(zip_code)
```

```python
len([k for k,v in permit_attr_hundreds_dict.items() if v is not None])
```

```python
def get_permits_for_zip(zip_code, permits=permits, prices=price_by_year):
    grouped_by_zip = permits[(permits.new_zip==zip_code) &
        (permits.ISSUE_DATE >= prices[zip_code].index.min()) &
        (permits.ISSUE_DATE <= prices[zip_code].index.max())].copy()
    grouped_by_zip['month_dt'] = grouped_by_zip.ISSUE_DATE.dt.strftime('%Y%m')
    grouped_by_zip = grouped_by_zip.groupby(['month_dt','PERMIT_TYPE'])['ID'].count().reset_index()
    reg_df = pd.DataFrame(dict(date=(grouped_by_zip.month_dt.unique()) )).set_index('date')
```

```python
grouped_60647 = permits[(permits.new_zip==60647) &
        (permits.ISSUE_DATE >= price_by_year[60647].index.min()) &
        (permits.ISSUE_DATE <= price_by_year[60647].index.max())].copy()
```

```python
grouped_60647['month_dt'] = grouped_60647.ISSUE_DATE.dt.strftime('%Y%m')
```

```python
grouped_60647 = grouped_60647.groupby(['month_dt','PERMIT_TYPE'])['ID'].count().reset_index()
```

```python
reg_df = pd.DataFrame(dict(date=grouped_60647.month_dt.unique())).set_index('date')
```

```python
grouped_60647.head()
```

```python
for perm_type,df in grouped_60647.groupby('PERMIT_TYPE'):
    cnt = df.set_index('month_dt')['ID']
    reg_df[perm_type] = cnt
```

```python
reg_df = reg_df.fillna(0)
```

```python
price_60647 = price_by_year[60647]
```

```python
reg_df['price'] = price_60647.values
```

```python
reg_df
```

```python
import statsmodels.api as sm
```

```python
y = reg_df.price
X = reg_df.drop('price', axis=1)
X = sm.add_constant(X)
```

```python
model = sm.OLS(y,X).fit()
```
