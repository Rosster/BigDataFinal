# Household Income & Poverty

```python
zips = set(permits.new_zip.unique())
```

```python
irs_data = []
for csv in tqdm_notebook(glob.glob('zip_irs_data\\*')):
    year = re.search(r'\d+',csv).group()
    year = '20' + year
    irs = pd.read_csv(csv)
    irs.columns = [str(c).lower() for c in irs.columns]
    grouped_irs = irs[irs.zipcode.isin(zips)].groupby('zipcode').sum()[['a00100','n1']]
    grouped_irs['income_per_filing'] = grouped_irs['a00100']/grouped_irs['n1']
    grouped_irs['year'] = year
    
    for level in range(1,7):
        grouped_irs[f'lvl_{level}'] = irs[irs.agi_stub==level].groupby('zipcode').sum().n1
    
    irs_data.append(grouped_irs)
```

```python
zip_income_data = pd.concat([df.reset_index() for df in irs_data]).rename({'a00100':'total_agi','n1':'n_tax_returns'}, axis=1)
```

```python
for c in zip_income_data.columns:
    if 'lvl_' in c:
        zip_income_data[c] = zip_income_data[c].apply(int)
```

```python
agg_home_dat = []
for csv in tqdm_notebook(glob.glob('housing_stock_data\\*')):
    year = re.search(r'\d+',csv).group()
    year = '20' + year
    home_dat = pd.read_csv(csv, skiprows=1)
    try:
        cols_of_interest = cols_of_interest_1
        rename_dict = rename_dict_1
        home_dat = home_dat[cols_of_interest]
        colgroup = 1
    except KeyError:
        cols_of_interest = cols_of_interest_2
        rename_dict = rename_dict_2
        try:
            home_dat = home_dat[cols_of_interest]
            colgroup = 2
        except KeyError:
            cols_of_interest = cols_of_interest_3
            rename_dict = rename_dict_3
            try:
                home_dat = home_dat[cols_of_interest]
                colgroup = 3
            except KeyError:
                cols_of_interest = cols_of_interest_4
                rename_dict = rename_dict_4
                home_dat = home_dat[cols_of_interest]
                colgroup = 4
    home_dat = home_dat.rename(rename_dict,axis=1)
    home_dat = home_dat[home_dat.zip_code.isin(zips)]
    if colgroup == 1:
        home_dat['built_2000_2009'] = home_dat.built_2000_2004 + home_dat.built_2005_present
        home_dat['built_2010_present'] = 0
        home_dat = home_dat.drop(['built_2000_2004','built_2005_present'], axis=1)
    elif colgroup == 4:
        home_dat['built_2010_present'] = home_dat.built_2010_2013 + home_dat.built_2014_present
        home_dat = home_dat.drop(['built_2010_2013','built_2014_present'], axis=1)
    home_dat['year'] = year
    agg_home_dat.append(home_dat)
```
